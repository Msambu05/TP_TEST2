```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/sambu/household_data.csv")
```

```python
df.columns
```

```python
# Plots using the basic infographic charts

# 1. Histogram

df = np.random.normal(loc = 20, scale = 5, size = 365)

plt.hist(df, bins = 25, color = 'skyblue', edgecolor = 'black')
plt.xlabel('Household_ID')
plt.ylabel('Population')
plt.title('Histogram of Population per Household')
plt.show()
```

```python
# 2. Pie chart

province_names = df['Province'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(province_names, labels = province_names.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen','yellow','red', 'grey', 'blue', 'orange', 'purple', 'brown'])
plt.title('Distribution of Provinces')
plt.show()
```

```python
# 3. Bar graph

df = np.random.normal(loc = 20, scale = 5, size = 10)

plt.figure(figsize=(8, 5)) 
plt.bar(np.arange(len(df)), df, color = 'lightblue', alpha = 1, label = 'Values_in_dataset')  
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Bar chart of Random Normal Data in the dataset')
plt.legend() 
plt.show()
```

```python
sns.set_theme()
sns.set_style("darkgrid")
```

```python
# Plots using Advanced charts

# 1. Boxplot

plt.figure(figsize=(10, 6))
sns.boxplot(x='Housing_Type', y='Household_ID', data=df)
plt.title('Box Plot: Household_ID by Housing_Type')
plt.xlabel('Housing_Type')
plt.ylabel('Household_ID')
plt.show()
```

```python
# 2. Pairplot

# Create categorical variables
df['Electricity_Access'] = df['Electricity_Access'].map({'Yes': 1, 'No': 0})
df['Water_Access'] = df['Water_Access'].map({'Yes': 1, 'No': 0})
df['Sanitation'] = df['Sanitation'].map({'Improved': 2, 'Good': 1, 'Poor': 0})

# Plot the pairplot
sns.pairplot(df[['Population', 'Electricity_Access', 'Water_Access', 'Sanitation']], hue='Electricity_Access')
plt.suptitle('Pairplot of Variables', y=1.02)
plt.show()
```

```python
# Summary Statistics

# Selecting Numeric columns
Numeric = df.select_dtypes(include=['number'])

# Summary for numeric columns
stats = Numeric.describe()

# Calculating median since it is not included in df.describe()
median = Numeric.median()

print("Summary Statistics (mean, std, min, 25%, 50%, 75%, max): ")
print(stats)
print("/Median: ")
print(median)
```
